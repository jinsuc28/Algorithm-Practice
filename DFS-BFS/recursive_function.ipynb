# 재귀함수 예제

```python
def recursive_function(i):
    if i ==1000:
        return
    print(i, '번째 재귀 함수에서', i+1,'번째 재귀 함수를 호출합니다')
    recursive_function(i+1)
    print(i, '번째 재귀 함수를 종료합니다')
```

```python
recursive_function(1)
```

# 유클리도 호제법 예제(필)

```python
#최대공약수
def gcd(a,b):
    if a%b==0:
        return b
    else:
        return gcd(b,a%b)
    
# 최소공배수
def lcm(a,b):
    return a*b*gcd(a,b)
    
```

```python
print(gcd(162,192))
print(lcm(162,192))
#순서는 상관 없다

# https://myjamong.tistory.com/138 참조
```
